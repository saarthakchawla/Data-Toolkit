def size_process(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace("M", " "))
        res = res*1000
        return str(res)
    elif str(item)[-1] == 'k':
        res = float(str(item).replace("k", ""))
        return str(res)
    else:
        return str(np.Nan)
